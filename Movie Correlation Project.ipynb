##########Lets Check For MIssing Data/Values#######\
i will create a for loop, that will loop through these columnsand within each one we will see if there is missing data

```python
for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {:.2f}%'.format(col, pct_missing))
```

```python
df.dtypes
```

```python
## fill missing values in 'budget' and 'gross' columns with zero

df['budget'].fillna(0, inplace = True)
df['gross'].fillna(0, inplace = True)
```

```python
## we will convert data types of columns from float to int
df['budget'] = df['budget'].astype('int64')
df['gross'] = df['gross'].astype('int64')
```

```python
df
```

```python
df['yearreleased'] = df['released'].str.extract(r'(\d{4})')
df
```

```python
df['yearcorrect'] = df['released'].str.extract(r'(\d{4})')
df
```

```python
# Drop any duplicates

df['company'].drop_duplicates().sort_values(ascending=False)
```

```python
df = df.sort_values(by=['gross'], inplace=False, ascending=False)
```

```python
df_numerized = df

for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes
        
df_numerized
```

```python
correlation_mat = df_numerized.corr()

corr_pairs = correlation_mat.unstack()

corr_pairs
```
